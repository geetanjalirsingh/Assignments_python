# Matplotlib assignment

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=pd.read_csv('movie_meta.csv')
```

```python
df1
```

```python
import numpy as np
m=np.mean(df1.actor_1_facebook_likes)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1.actor_1_facebook_likes[pd.isna(df1.actor_1_facebook_likes)]=m
```

```python
m1=np.mean(df1.num_critic_for_reviews)
df1.num_critic_for_reviews[pd.isna(df1.num_critic_for_reviews)]=m1
```

```python
plt.figure(figsize=(10,20))
plt.scatter(df1.num_critic_for_reviews, df1.actor_1_facebook_likes, color='red', label ='actor 1 likes')

plt.scatter(df1.num_critic_for_reviews, df1.actor_2_facebook_likes, color='green', label ='actor 2 likes')

plt.scatter(df1.num_critic_for_reviews, df1.actor_3_facebook_likes, color='blue', label ='actor 3 likes')

plt.legend()
plt.xlabel('Critic Reviews')
plt.ylabel('facebook likes')
plt.title('Critic Reviews of movie vs facebook likes')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.hist(df1.genres, bins=30)
plt.xlabel('Genres')
plt.ylabel('frequency')
plt.title('Generes graph')
plt.show()
```

```python
df1.language.unique()
```

```python
l=df1.language.value_counts()
```

```python
plt.figure(figsize=(10,10))
plt.pie(labels=l.index, x=l)
plt.show()
```
